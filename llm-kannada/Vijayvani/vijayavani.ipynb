```python
def get_driver():
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--disable-extensions")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--disable-dev-shm-usage")
    # chrome_options.add_argument("--disable-images")  # Disable images
    # chrome_options.add_argument("--disable-javascript")  # Disable JS if possible
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
    driver = webdriver.Chrome(options=chrome_options)
    return driver
```

```python
links = []
for i in range(start_page, end_page + 1):
    print("Page number: ", i)
    try:
        driver = get_driver()
        driver.get(base_url + 'page/' + str(i))
        driver.minimize_window()

        # Use WebDriverWait to wait for the element to be present
        wait = WebDriverWait(driver, 10)
        div1 = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[class="blog-content"]')))
        
        titles = div1[0].find_elements(By.CSS_SELECTOR, '[class="entry-title"]')
        if len(titles) == 0:
            print("Unable to find article titles on page:", i)
            continue
        for title in titles:
            try:
                article_url = title.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
                print("   Article URL:", article_url)
                links.append(article_url)
            except Exception as e:
                print(f"Error extracting URL from title on page {i}: {e}")
        print("   Links collected:", len(links))  

    except Exception as e:
        print(f"Error processing page {i}: {e}")

    finally:
        driver.quit()

    time.sleep(random.randint(1, 5))

print("Total links collected:", len(links))
```
