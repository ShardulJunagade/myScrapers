```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
import random

# karnataka news
base_url='https://www.vijayavani.net/category/%e0%b2%b8%e0%b2%ae%e0%b2%b8%e0%b3%8d%e0%b2%a4-%e0%b2%95%e0%b2%b0%e0%b3%8d%e0%b2%a8%e0%b2%be%e0%b2%9f%e0%b2%95/'
start_page = 1
end_page = 4565
```

```python
def get_driver():
    chrome_options = Options()
    # chrome_options.add_argument("--headless")
    chrome_options.add_argument("--disable-extensions")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--disable-images")  # Disable images
    chrome_options.add_argument("--disable-javascript")  # Disable JS if possible
    driver = webdriver.Chrome(options=chrome_options)
    return driver
```

```python
driver = get_driver()
driver.get(base_url)
div1 = driver.find_elements(By.CSS_SELECTOR, '[class="blog-content"]')
if len(div1)==0:
    print("Some problem finding the grid")
# print(len(div1))
links = []

titles = div1[0].find_elements(By.CSS_SELECTOR, '[class="entry-title"]')
if len(titles)==0:
    print("Some problem finding the titles")
for title in titles:
    article_url = title.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
    print(article_url)
    links.append(article_url)
driver.quit()
```

```python
links = []
for i in range(3350, 4001):
    print("Page number: ", i)
    try:
        driver = get_driver()
        driver.get(base_url + 'page/' + str(i))
        driver.minimize_window()
        div1 = driver.find_elements(By.CSS_SELECTOR, '[class="blog-content"]')
        # print(len(div1))
        if len(div1) == 0:
            print("Unable to find blog grid on page:", i)
            continue
    
        titles = div1[0].find_elements(By.CSS_SELECTOR, '[class="entry-title"]')
        if len(titles) == 0:
            print("Unable to find article titles on page:", i)
            continue
        for title in titles:
            try:
                article_url = title.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
                print("   Article URL:", article_url)
                links.append(article_url)
            except Exception as e:
                print(f"Error extracting URL from title on page {i}: {e}")

    except Exception as e:
        print(f"Error processing page {i}: {e}")

    finally:
        driver.quit()

    time.sleep(random.randint(1, 5))

print("Total links collected:", len(links))
```

```python
df = pd.DataFrame(columns=["url"])
for i in range(len(links)):
    df.loc[len(df)] = [links[i]]

df
```
